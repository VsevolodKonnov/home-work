```python
def DigitCountSum(K):
    count=0
    sum=0
    while K>0:
        d=K%10
        if d>0:
            count += 1
            sum += d
        K=K//10
    return (count, sum)

x,y=DigitCountSum(1578)
print(x)
print(y)
```

```python
def factorial(k):
    if k == 0 or k == 1:
        return 1
    result = 1
    for i in range(2, k + 1):
        result *= i
    return result

def sum_factorials(n):
    if n % 2 != 0 or n < 2 or n > 20:
        return Error
    total = 0
    for k in range(2, n + 1, 2):
        total += factorial(k)
    return total

result = sum_factorials(6)
print(result)
```
